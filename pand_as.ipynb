### Pandas

```Pandas``` is a Python library used for working with data sets.

It has functions for analyzing, cleaning, exploring, and manipulating data.

The name "Pandas" has a reference to both "Panel Data", and "Python Data Analysis" and was created by Wes McKinney in 2008.

### Why Use Pandas?
```Pandas``` allows us to analyze big data and make conclusions based on statistical theories.

Pandas can clean messy data sets, and make them readable and relevant.

Relevant data is very important in data science.

````python
print(df['price'].min())#get the minimum value of the column 'sell_price'
print(df['price'].max())#get the maximum value of the column 'sell_price'
print(df['price'].mean())#get the mean value of the column 'sell_price' 
print(df['price'].median())#get the median value of the column 'sell_price'
print(df['price'].std())#get the standard deviation of the column 'sell_price'
print(df['price'].unique())#get the unique values of the column 'sell_price'
print(df['price'].nunique())#get the number of unique values of the column 'sell_price'
print(df['price'].describe())#get the descriptive statistics of the column 'sell_price'
print(df['city'].value_counts())#get the count of each unique value of the column 'sell_price'
print(df.columns)
````
